```python
from PIL import Image
import numpy as np

def niveau_gris_moyen(image_path):
    # Ouvrir l'image
    image = Image.open(image_path)

    # Convertir l'image en niveaux de gris
    image_gray = image.convert('L')

    # Convertir l'image en tableau numpy
    image_np = np.array(image_gray)

    # Calculer le niveau de gris moyen
    mean_gray = np.mean(image_np)

    return mean_gray

# Chemin de base vers le rpertoire contenant les sous-dossiers d'images
base_path = '/content/PBC_dataset_normal_DIB'


# Fonction pour construire le chemin complet de l'image  partir du nom du fichier
def get_image_path(nom_fichier):
    for root, dirs, files in os.walk(base_path):
        if nom_fichier in files:
            return os.path.join(root, nom_fichier)
    return None

# Ajouter une nouvelle colonne avec le nombre de gros contours
df['Niveau de gris'] = df['Nom fichier'].apply(lambda nom_fichier: niveau_gris_moyen(get_image_path(nom_fichier)) if get_image_path(nom_fichier) else None)

# Afficher la DataFrame
print(df)
```

```python
import numpy as np
from PIL import Image
from scipy import ndimage
def decouper_centre(image_np, proportion=0.5):
    """
    Dcoupe la partie centrale de l'image en fonction de la proportion spcifie.
    La proportion doit tre comprise entre 0 et 1.
    """
    h, w = image_np.shape
    centre_h = int(h * proportion)
    centre_w = int(w * proportion)
    debut_h = (h - centre_h) // 2
    debut_w = (w - centre_w) // 2
    return image_np[debut_h:debut_h + centre_h, debut_w:debut_w + centre_w]

def compter_contours(image_path, seuil=130, proportion=0.5):
    # Ouvrir l'image et convertir en niveaux de gris
    image = Image.open(image_path).convert('L')

    # Convertir l'image en tableau numpy
    image_np = np.array(image)

    # Dcouper la partie centrale de l'image
    image_centre = decouper_centre(image_np, proportion)

    # Appliquer le seuillage
    image_seuillee = (image_centre > seuil).astype(np.uint8) * 255

    # Trouver les contours en utilisant les diffrences de gradients
    sx = ndimage.sobel(image_seuillee, axis=0, mode='constant')
    sy = ndimage.sobel(image_seuillee, axis=1, mode='constant')
    sobel = np.hypot(sx, sy)

    # Binariser le rsultat du filtre Sobel
    contours = (sobel > 0).astype(np.uint8)

    # tiqueter les objets (contours)
    labeled_array, num_features = ndimage.label(contours)

    return num_features

# Chemin de base vers le rpertoire contenant les sous-dossiers d'images
base_path = '/content/PBC_dataset_normal_DIB'


# Fonction pour construire le chemin complet de l'image  partir du nom du fichier
def get_image_path(nom_fichier):
    for root, dirs, files in os.walk(base_path):
        if nom_fichier in files:
            return os.path.join(root, nom_fichier)
    return None

# Ajouter une nouvelle colonne avec le nombre de gros contours
df['Petits contours'] = df['Nom fichier'].apply(lambda nom_fichier: compter_contours(get_image_path(nom_fichier)) if get_image_path(nom_fichier) else None)

# Afficher la DataFrame
df
```

```python
from PIL import Image
import numpy as np
from scipy import ndimage
import pandas as pd
import os

# Fonction pour compter les gros contours avec des messages de dbogage dtaills
def compter_gros_contours(image_path, seuil=130, proportion=0.5, aire_min=100):
    try:
        print(f"Tentative d'ouverture de l'image : {image_path}")
        # Ouvrir l'image et convertir en niveaux de gris
        image = Image.open(image_path).convert('L')

        # Convertir l'image en tableau numpy
        image_np = np.array(image)

        # Dcouper la partie centrale de l'image
        h, w = image_np.shape
        centre_h = int(h * proportion)
        centre_w = int(w * proportion)
        debut_h = (h - centre_h) // 2
        debut_w = (w - centre_w) // 2
        image_centre = image_np[debut_h:debut_h + centre_h, debut_w:debut_w + centre_w]

        # Appliquer le seuillage
        image_seuillee = (image_centre > seuil).astype(np.uint8) * 255

        # Trouver les contours en utilisant les diffrences de gradients
        sx = ndimage.sobel(image_seuillee, axis=0, mode='constant')
        sy = ndimage.sobel(image_seuillee, axis=1, mode='constant')
        sobel = np.hypot(sx, sy)

        # Binariser le rsultat du filtre Sobel
        contours = (sobel > 0).astype(np.uint8)

        # tiqueter les objets (contours)
        labeled_array, num_features = ndimage.label(contours)

        # Filtrer les contours par leur aire
        areas = ndimage.sum(contours, labeled_array, range(1, num_features + 1))
        gros_contours = areas > aire_min
        nombre_gros_contours = np.sum(gros_contours)

        return nombre_gros_contours
    except Exception as e:
        print(f"Une erreur s'est produite pour l'image {image_path} : {e}")
        return None

# Chemin de base vers le rpertoire contenant les sous-dossiers d'images
base_path = '/content/PBC_dataset_normal_DIB'


# Fonction pour construire le chemin complet de l'image  partir du nom du fichier
def get_image_path(nom_fichier):
    for root, dirs, files in os.walk(base_path):
        if nom_fichier in files:
            return os.path.join(root, nom_fichier)
    return None

# Ajouter une nouvelle colonne avec le nombre de gros contours
df['Gros contours'] = df['Nom fichier'].apply(lambda nom_fichier: compter_gros_contours(get_image_path(nom_fichier)) if get_image_path(nom_fichier) else None)

# Afficher la DataFrame
print(df)
```

```python
def compter_pixels_sup_seuil(image_path, seuil=130):
    # Ouvrir l'image en niveaux de gris
    image = Image.open(image_path).convert('L')

    # Convertir l'image en tableau numpy
    image_np = np.array(image)

    # Appliquer le seuillage (si l'image n'est pas encore seuille)
    image_seuillee = (image_np > seuil).astype(np.uint8) * 255

    # Compter le nombre de pixels au-dessus du seuil
    nombre_pixels_sup_seuil = np.sum(image_seuillee > 0)

    return nombre_pixels_sup_seuil

# Chemin de base vers le rpertoire contenant les sous-dossiers d'images
base_path = '/content/PBC_dataset_normal_DIB'


# Fonction pour construire le chemin complet de l'image  partir du nom du fichier
def get_image_path(nom_fichier):
    for root, dirs, files in os.walk(base_path):
        if nom_fichier in files:
            return os.path.join(root, nom_fichier)
    return None

# Ajouter une nouvelle colonne avec le nombre de gros contours
df['Nombre pixels'] = df['Nom fichier'].apply(lambda nom_fichier: compter_pixels_sup_seuil(get_image_path(nom_fichier)) if get_image_path(nom_fichier) else None)

# Afficher la DataFrame
print(df)
```

```python
X = df[['Gros contours', 'Nombre pixels', 'Petits contours', 'Niveau de gris']]
y = df['Code globule']

# Crer et entraner l'arbre de dcision
clf = DecisionTreeClassifier(criterion='gini', max_depth=3)
clf.fit(X, y)

# Extraire des informations sur l'arbre
n_nodes = clf.tree_.node_count
children_left = clf.tree_.children_left
children_right = clf.tree_.children_right
feature = clf.tree_.feature
threshold = clf.tree_.threshold
value = clf.tree_.value

# Fonction pour imprimer l'information de chaque nud
def print_node_info(node):
    if children_left[node] == children_right[node]:  # Nud feuille
        print(f"Node {node} is a leaf node.")
        print(f"Value: {value[node]}")
    else:
        print(f"Node {node}:")
        print(f"  Feature: {feature[node]}")
        print(f"  Threshold: {threshold[node]}")
        print(f"  Left child: {children_left[node]}")
        print(f"  Right child: {children_right[node]}")
        print(f"  Value: {value[node]}")

# Parcourir tous les nuds et imprimer leurs informations
for node in range(n_nodes):
    print_node_info(node)
```
