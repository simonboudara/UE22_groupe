```python
from PIL import Image

# Ouvrir l'image
image = Image.open("/Users/simonboudara/Cours_info/Projet_machine_learning/dataset/PBC_dataset_normal_DIB/basophil/BA_1223.jpg")

# Convertir l'image en niveaux de gris
image_grayscale = image.convert('L')

# Appliquer le seuillage sur l'image en niveaux de gris
seuil = 110
image_seuillee = image_grayscale.point(lambda x: x if x < seuil else 255)

# Sparer les canaux RGB de l'image originale
r, g, b = image.split()

# Appliquer le seuillage sur chaque canal
r_seuille = r.point(lambda p: p if p < seuil else 255)
g_seuille = g.point(lambda p: p if p < seuil else 255)
b_seuille = b.point(lambda p: p if p < seuil else 255)

# Fusionner les canaux seuills avec les canaux de l'image originale
image_seuillee = Image.merge('RGB', (r_seuille, g_seuille, b_seuille))

# Afficher l'image seuille
plt.imshow(image_seuillee)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Charger l'image en niveaux de gris
image = plt.imread('/Users/simonboudara/Documents/tmptf3wuxrg.PNG')

# Convertir l'image en niveaux de gris si ncessaire
if len(image.shape) == 3:
    image_grayscale = np.mean(image, axis=2)
else:
    image_grayscale = image

# Appliquer un filtre de Sobel pour dtecter les contours
dx = np.gradient(image_grayscale, axis=1)
dy = np.gradient(image_grayscale, axis=0)
contours = np.sqrt(dx**2 + dy**2)

# Binariser l'image en utilisant un seuil adaptatif
seuil = np.mean(contours) + 2 * np.std(contours)
contours_binaires = contours > seuil

# Trouver les contours des rgions connectes
contours_connectes, _ = cv2.findContours(contours_binaires.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Afficher l'image avec les contours dtects
plt.figure(figsize=(8, 6))
plt.imshow(image_grayscale, cmap='gray')
for contour in contours_connectes:
    for point in contour:
        plt.plot(point[0][0], point[0][1], 'ro')
plt.axis('off')
plt.show()

# Afficher le nombre de contours dtects
print("Nombre de contours dtects:", len(contours_connectes))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Charger l'image en niveaux de gris
image = plt.imread('/Users/simonboudara/Documents/tmptf3wuxrg.PNG')

# Convertir l'image en niveaux de gris si ncessaire
if len(image.shape) == 3:
    image_grayscale = np.mean(image, axis=2)
else:
    image_grayscale = image

# Appliquer un filtre de Sobel pour dtecter les contours
dx = np.gradient(image_grayscale, axis=1)
dy = np.gradient(image_grayscale, axis=0)
contours = np.sqrt(dx**2 + dy**2)

# Binariser l'image en utilisant un seuil adaptatif
seuil = np.mean(contours) + 2 * np.std(contours)
contours_binaires = contours > seuil

# Appliquer une opration de dilatation pour agrandir les contours
kernel = np.ones((5, 5), np.uint8)  # kernel de dilatation
contours_binaires_dilate = cv2.dilate(contours_binaires.astype(np.uint8), kernel, iterations=1)

# Trouver les contours des rgions connectes
contours_connectes, _ = cv2.findContours(contours_binaires_dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Afficher l'image avec les contours dtects
plt.figure(figsize=(8, 6))
plt.imshow(image_grayscale, cmap='gray')
for contour in contours_connectes:
    for point in contour:
        plt.plot(point[0][0], point[0][1], 'ro')
plt.axis('off')
plt.show()

# Afficher le nombre de contours dtects
print("Nombre de contours dtects:", len(contours_connectes))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans

# Charger l'image
image = cv2.imread('/Users/simonboudara/Documents/tmptf3wuxrg.PNG')

# Convertir l'image de BGR  RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Redimensionner l'image si ncessaire pour acclrer le processus
# image = cv2.resize(image, (300, 300))

# Convertir l'image en une matrice 2D de pixels RGB
pixels = image.reshape((-1, 3))

# Dfinir le nombre de clusters (zones de couleurs diffrentes)
nombre_clusters = 5  # Vous pouvez ajuster ce paramtre selon votre image

# Appliquer l'algorithme de K-Means pour regrouper les pixels en clusters
kmeans = KMeans(n_clusters=nombre_clusters)
kmeans.fit(pixels)

# Obtenir les tiquettes de cluster pour chaque pixel
labels = kmeans.labels_

# Nombre de zones de couleurs diffrentes
nombre_zones = len(np.unique(labels))

# Afficher l'image originale avec les clusters colors
cluster_centers = kmeans.cluster_centers_.astype(int)
segmented_image = cluster_centers[labels].reshape(image.shape)

plt.figure(figsize=(8, 6))
plt.imshow(segmented_image)
plt.axis('off')
plt.title("Image segmente en {} zones de couleurs diffrentes".format(nombre_zones))
plt.show()

print("Nombre de zones de couleurs diffrentes dans l'image:", nombre_zones)
```

```python
import numpy as np
import cv2

# Charger l'image en couleurs
image = cv2.imread('/Users/simonboudara/Documents/tmptf3wuxrg.PNG')

# Convertir l'image en niveaux de gris
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Appliquer un seuillage adaptatif pour obtenir un masque binaire
_, mask = cv2.threshold(image_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Effectuer une opration de fermeture pour liminer les petits trous dans le masque
kernel = np.ones((5,5), np.uint8)
mask_closed = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# Trouver les contours des zones significatives
contours, _ = cv2.findContours(mask_closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dessiner les contours sur l'image originale
result = image.copy()
cv2.drawContours(result, contours, -1, (0, 255, 0), 2)

# Afficher l'image rsultante
cv2.imshow('Result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Afficher le nombre de zones significatives
print("Nombre de zones significatives:", len(contours))
```
